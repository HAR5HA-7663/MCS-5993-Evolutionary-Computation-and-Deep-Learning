# **matplotlib** basics - 2D contour plots

* https://matplotlib.org/
* https://matplotlib.org/2.0.2/gallery.html (gallery)

```python
import matplotlib.pyplot as plt
import numpy as np
low = -5
high = 5
xlist = np.linspace(low, high, 100)
ylist = np.linspace(low, high, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = X*X + 2*Y*Y
```

```python
cp = plt.contour(X, Y, Z)
plt.clabel(cp)
plt.title('Contour plot of X^2 + 2Y^2')
plt.xlabel('x0')
plt.ylabel('x1')
plt.show()
```

```python
ax = plt.subplot() # whole figure and an axis (the choice of the name axis is wrong??)
cp = plt.contour(X, Y, Z) # X, Y are from meshgrid
ax.clabel(cp)
ax.set_title("x^2 + 2Y^2")
ax.set_xlabel('x0')
ax.set_ylabel('x1')
plt.show()
```

```python
Z1 = 5*X*X + Y*Y
Z2 = X*X - Y*Y
fig, axes = plt.subplots(2,2)

cp0 = axes[0][0].contour(X, Y, Z)
cp1 = axes[0][1].contour(X, Y, Z1)
cp2 = axes[1][0].contour(X, Y, Z2)
axes[1][1].plot(X,Y)

axes[0][0].clabel(cp0)
axes[0][0].set_title("X^2 + 2Y^2")
axes[0][0].set_ylabel('x1')
axes[0][0].set_xlabel('x0')

axes[0][1].clabel(cp1)
axes[0][1].set_title("5X^2 + Y^2")
axes[0][1].set_xlabel('x0')
axes[0][1].set_ylabel('x1')

axes[1][0].clabel(cp2)
axes[1][0].set_title("X^2 - Y^2")
axes[1][0].set_xlabel('x0')
axes[1][0].set_ylabel('x1')

fig.tight_layout() # utomatically maintains correct spacing among subplots.
plt.show()
```
