```python
x = np.linspace(0, 10, 10)
y = x*x
print(x)
print(y)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-np.pi, np.pi, 0.01) # Grid of 0.01 spacing from -pi to pi
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
def func(x):
    return x**2 - 1
low = -5
high = 5
X = np.arange(low, high, 0.5) # Grid of 0.5 spacing from low to high
Y = func(X) # Evaluate function on xvals
```

```python
print(X)
print(Y)
```

```python
plt.plot(X, Y) # Create line plot with yvals against xvals
plt.show() # Show the figure
```

```python
plt.plot(X, Y, '-v', color="blue") # Create line plot with yvals against xvals
plt.show() # Show the figure
```

#### Moving Spines

```python
X1 = np.arange(-np.pi, np.pi, 0.01) # Grid of 0.01 spacing from -pi to pi
Y1 = np.sin(x)
ax = plt.gca()         # Get Current Axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.plot(X1, Y1) # Create line plot with yvals against xvals
plt.show() # Show the figure
```

```python
def func(x):
    return x**2 - 2
low = -5
high = 5
X = np.arange(low, high, 0.5) # Grid of 0.5 spacing from low to high
Y = func(X) # Evaluate function on xvals
ax = plt.gca()         # Get Current Axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
#ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.plot(X, Y) # Create line plot with yvals against xvals
plt.show() # Show the figure
```

```python
#--------- A hint for a possible assignment (using Lists) -----
xp = []
yp = []

for x in np.arange(0.0, 1.05, 0.05): # cannot use range since start and stop must be int
  for y in np.arange(0.0, 1.05, 0.05):
    xp.append(x)
    yp.append(y)

# -------- PLOT the graph -----
plt.plot(xp, yp, 'k.') # to plot 4 points in x. k is for black
plt.show() # Show the figure
```

## Scatter Plots with plt.scatter
* More powerful than plt.plot
* Quite similar to plt.plot
* The properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data

```python
plt.scatter(X, Y, marker='s')
plt.show()
```

```python
plt.scatter(X[0], Y[0], marker='s')
plt.scatter(X[1:-1], Y[1:-1], marker='.')
plt.scatter(X[-1], Y[-1], marker='o')
plt.show()
```

```python
rg = np.random.RandomState(0)
X = rg.randn(500)  # 50 points. What is randn? standard Normal distribution
Y = rg.randn(500)
Colors = rg.rand(500)
#Sizes = 100 * rg.rand(500)

plt.scatter(X, Y, c=Colors, #s=Sizes,
            alpha=0.8,      # 0.0 transparent through 1.0 opaque
            cmap='viridis') # registered colormap name
plt.colorbar();  # show color scale bar
```
