## (Solution 0) Testing with one neuron (perceptron) - cannot be solved. Proven by Dr. Papert

```
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

X1= np.array([[0.1,0],[0.1,1],[1,0.1],[0.9,1]])

model = Sequential([
    Dense(1, input_dim=2, activation='sigmoid') # 'tanh'
])
sgd = optimizers.SGD(learning_rate=0.1)
model.compile(loss='mean_squared_error', # https://keras.io/losses/
              optimizer=sgd)             # https://keras.io/optimizers/

model.fit(X, y, batch_size=1, epochs=1000, verbose=0)
# batch size: Number of samples per gradient update.
# An epoch is an iteration over the entire X and y data provided
# total number of gradient updates: 4/1 * 100 = 400
```

```
print(model.predict(X).round()) # does not work...
print(model.predict(X1).round()) # does not work...
```

##(Solution 1) Simplest model using 2 hidden neurons and SGD optimizer, tanh activation function
<image src=https://www.robofest.net/DL/XOR1.jpg width=300>

```
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = Sequential([
    Dense(2, input_dim=2, activation='tanh'),
    Dense(1, activation='tanh')
])

model.compile(
    loss='mean_squared_error',
    optimizer=optimizers.SGD(learning_rate=0.1), # stochastic gradient decent
    #optimizer=optimizers.RMSprop(learning_rate=0.01),
    metrics=['binary_accuracy']
)
model.fit(X, y, batch_size=1, epochs=2000, verbose=0)
```

```
print(model.predict(X))
print(model.predict(X).round())

X1 = np.array([[0,0.1],[0,1],[0.9,0],[1,1]]) # new test data
print(model.predict(X1).round())
```

```
for layer in model.layers:
    print(layer.get_weights())
```

##(Solution 1a) Simplest model using 2 hidden neurons and SGD optimizer, sigmoid activation function
<image src=https://www.robofest.net/DL/XOR_sigmoid.png width=300>

```
# using sigmoid function
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import numpy as np

X = np.array([[0,0],[0.1,0.1],[0,1],[0.1,0.9],[1,0],[0.9,0.1],[0.9, 0.9],[1,1]])
y = np.array([[0],  [0],      [1],  [1],      [1],  [1],      [0],       [0]])

model = Sequential([
    Dense(2, input_dim=2, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])

model.compile(
    loss='mean_squared_error',
    optimizer=optimizers.SGD(learning_rate=0.1), # stochastic gradient decent
    #optimizer=optimizers.RMSprop(learning_rate=0.01),
    metrics=['binary_accuracy']
)
model.fit(X, y, batch_size=1, epochs=2000, verbose=0)
```

```
print(model.predict(X))
print(model.predict(X).round())

X1 = np.array([[0,0.1],[0,1],[0.9,0],[1,1]]) # new test data
print(model.predict(X1).round())
```

## (Solution 2) with 16 hidden neurons and adam optimizer
<image src=https://www.robofest.net/DL/XOR16.JPG width=300>

```
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers

# the four different states of the XOR gate
X = np.array([[0,0],[0,1],[1,0],[1,1]], "float32") #training data

# the four expected results in the same order
y = np.array([[0],[1],[1],[0]], "float32") # target labels

m2 = Sequential([
    Dense(16, input_dim=2, activation='relu'), # relu worked
    #Dense(16, input_dim=2, activation='sigmoid'), #tanh or sigmod seems not working
    Dense(1, activation='sigmoid')
])

m2.compile(
    loss='mean_squared_error',
    optimizer=optimizers.Adam(),
    metrics=['binary_accuracy'] # metrics is to monitor and display.
    # https://machinelearningmastery.com/custom-metrics-deep-learning-keras-python/
)
m2.fit(X, y, batch_size=1, epochs=2000, verbose=0)
```

```
print (m2.predict(X))
print (m2.predict(X).round())

test_data = np.array([[0,0.01],[0,0.9],[1,0],[1,1]], "float32")
print (m2.predict(test_data).round())
```

```
first_layer_weights = m2.layers[0].get_weights()[0]
first_layer_biases  = m2.layers[0].get_weights()[1]
second_layer_weights = m2.layers[1].get_weights()[0]
second_layer_biases  = m2.layers[1].get_weights()[1]

print(first_layer_weights)
print(first_layer_biases)
print(second_layer_weights)
print(second_layer_biases)
```

## Solution 3: 2 hidden layers

```
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers

# the four different states of the XOR gate
X = np.array([[0,0],[0,1],[1,0],[1,1]], "float32") #training data

# the four expected results in the same order
y = np.array([[0],[1],[1],[0]], "float32") # target labels

m2 = Sequential([
    Dense(12, input_dim=2, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])

m2.compile(
    loss='mean_squared_error',
    optimizer=optimizers.SGD(learning_rate=0.01),
    metrics=['binary_accuracy'] # metrics is to monitor and display.
    # https://machinelearningmastery.com/custom-metrics-deep-learning-keras-python/
)
m2.fit(X, y, batch_size=1, epochs=2000, verbose=0)
```

```
print (m2.predict(X).round())

test_data = np.array([[0,0.01],[0,0.9],[1,0],[1,1]], "float32")
print (m2.predict(test_data).round())
```

```
print(m2.layers[0].get_weights()[0]) # 1st layer w
print(m2.layers[0].get_weights()[1]) # 1st layer bias
print(m2.layers[1].get_weights()[0]) # 2nd layer w
print(m2.layers[1].get_weights()[1]) # 2nd layer b
print(m2.layers[2].get_weights()[0]) # 3rd layer w
print(m2.layers[2].get_weights()[1]) # 3rd layer b
```

```
for layer in m2.layers:
    print(layer.get_weights())
```
