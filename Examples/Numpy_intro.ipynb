```python
import numpy as np           # people use "np"
celcius = [20, 21, 22, 23, 34, 36]
fahrenheight = [ x*9/5 + 32 for x in celcius ]
print(fahrenheight)
```

```python
# better way using numpy
C = np.array(celcius) # C is not a list. We call it as an array
F = C*9/5 + 32
print(F)
```

Are "celcius" and "C" same? What is the difference? Note there is no comma between elements. C is not a list. We referred to C as an array. The internal type is "ndarray" or to be even more precise "C is an instance of the class numpy.ndarray":

```python
type(C)
```

```python
C2 = 5/9*(F-32)
print(C2)
```

```python
print(np.sin(F))
```

```python
print(C+F)
print(C*F)
```

### Indexing Numpy arrays (same as lists)
* First index is zero, not one
* The end value is not inclusive [0:n)

```python
import numpy as np
a = np.array([0,1,2,3,4,5])
print(a[1:-1]) # -1: excluding the last item
```

## Python Slicing for string/list/numpy/tuple
* https://www.geeksforgeeks.org/python-list-slicing/
* https://stackoverflow.com/questions/509211/understanding-slicing
* https://stackoverflow.com/questions/31633635/what-is-the-meaning-of-inta-1-in-python

```python
import numpy as np
a = np.array([2, 3, 4, 5, 6, 7, 999])
print(a[0])
print(a[:3])  # array slicing, from index 0 to index 2
print(a[3:])  # from index 3 till the end
print(a[2:4]) # from index 2 to 3
```

```python
C = np.array([0,1,2,3,4,5])
print("Original array", C)
print(C[0:2])
print(C[0:])
print(C[:])
print(C[:3])
print(C[1:2])
print(C[:-1])  # excluding last item
print(C[1:-1]) # excluding first & last item
print(C[1:-2]) # excluding last 2 items
print(C[-1])   # last item
```

```python
# how to get the last item
print(C[-1])
print(C[-1:])
print(C[-2]) # second last item
```

## Numpy.where()
returns the indices of elements in an input array where the given condition is satisfied

```python
import numpy as np

# a is an array of integers.
a = np.array([1, 2, 3, 4, 5, 6])

print ('Indices of elements < 3')
i = np.where(a<3)
print(i)
print(i[0])
b = a[i[0]]
print(b)
```

```python
b = np.array([1, 2, 3, 4, 9, 6])
i = np.where(a != b)
print(i)
```

```python
# Type A Question

aa = np.array([[3,0,1,0], [0,0,0,1], [0,1,1,2]])
bb = np.array([[3,0,1,0], [0,0,3,0], [0,1,1,2]])
print("A = ", np.argmax(aa, axis=1))
print("B = ", np.argmax(bb, axis=1))
i = np.where(np.argmax(aa, axis=1) != np.argmax(bb, axis=1))[0]
print("C = ", i)
```

```python
# Type B Question

aa = np.array([ [0,0,0,1], [0,1,1,2], [3,0,1,0]])
bb = np.array([ [0,0,3,0], [0,1,1,2], [3,0,1,0]])
print("A = ", np.argmax(aa, axis=1))
print("B = ", np.argmax(bb, axis=1))
i = np.where(np.argmax(aa, axis=1) != np.argmax(bb, axis=1))[0]
print("C = ", i)
```
